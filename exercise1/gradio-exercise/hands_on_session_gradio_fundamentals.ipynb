### **Chatbots**

- Gradio includes a high-level class, `gr.ChatInterface`.
- The `gr.ChatInterface` class also wraps a function but this function must have a specific signature.
- The function should take two arguments: `message` and then `history` (the arguments can be named anything, but must be in this order)

  - `message`: a str representing the users input
  - `history`: a list of list representing the conversations up until that point. Each inner list consists of two str representing a pair: [user input, bot response].

Your function should return a single string response, which is the bots response to the particular user input message.

Other than that, `gr.ChatInterface` has no required parameters (though several are available for customization of the UI).

### **Managing API Access in Your Gradio Chatbot Application**

- When you deploy a Gradio chatbot, it includes a feature that might not always align with your privacy or security policies:  "Use via API" link in the application footer.

 image.png

 - This link allows anyone with access to your Gradio interface to see how to use your application programmatically, potentially increasing the risk of unwanted access or use.

#### **Why Consider Hiding the "Use via API" Option?**
- **Security:**
  - Limiting how your application can be accessed or used is a fundamental security practice.
  - Exposing the API access points can make your application a target for unauthorized use.
- **Control:**
  - You may want to control who can access your application and how they can use it.
  - By not publicly sharing the API details, you maintain more control over your application's use.
- **Privacy:**
  - Depending on the nature of your chatbot, you might be handling sensitive information.
  - Keeping the API access under wraps helps in safeguarding this information.

#### **How to Hide the "Use via API" Footer**
- If you decide that hiding the API access information is the right step for you, here is a simple CSS trick you can use to remove the footer that displays the "Use via API" link in your Gradio ChatInterface:

```Python
# UI Elements (Frontend)
with gr.Blocks(css="footer{display:none !important}") as demo:
    with gr.Row():
        create_logo()
        create_title("IABG's ChatGPT")
    # Chat Interface
    gr.ChatInterface(predict)

demo.queue().launch(share=True)
```

```Python
import gradio as gr
import openai
openai.api_key = "" ## Add Your API key

def create_logo(filename=None):
    return gr.Image(
              sources=filename or "upload" or "iabg_logo.png",
              width=5,
              height=75,
              scale=0,
              show_label=False,
              show_download_button=False,
          )

def create_title(title):
    return gr.Markdown(
              f"<h1>{title}</h1>" +
              """
                  <style>
                      .gradio-container-4-21-0 .prose h1 {
                          text-align: center;
                          margin-top : 10px;
                          padding-top : 10px;
                          margin-left : 300px;
                          font-size: 2em;
                      }
                  </style>
              """
          )
    
# Logic (backend)
def predict(prompt, history):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=0,
        stream=True,
    )
    #### ADD Streaming ####
    partial_message = ""
    for chunk in response:
        if chunk:
          delta = getattr(chunk.choices[0], 'delta', {})
          # Safely get the 'content' from 'delta'
          content = delta.get('content', "") if delta else ""
          partial_message += (content or "")
          yield partial_message

# UI Elements (Frontend)
with gr.Blocks(css="footer{display:none !important}") as demo:
    with gr.Row():
        create_logo()
        create_title("IABG's ChatGPT")
    # Chat Interface
    gr.ChatInterface(predict)

# Launch the gradio app for 72 hours
demo.queue().launch()

```
